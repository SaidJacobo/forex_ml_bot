```python
import pandas as pd

INITIAL_CASH = 10_000


equity_path = 'C:/Users/SaidJ/OneDrive/Documentos/projects/forex_ml_bot/forex_ml_bot/backtesting_pipeline/strategies/ShortIBS/preliminar_analysis/BABA_16385/equity.csv'
prices_path = 'C:/Users/SaidJ/OneDrive/Documentos/projects/forex_ml_bot/forex_ml_bot/backtesting_pipeline/get_data/data/BABA_16385_2021-01-01_2024-11-01.csv'

equity = pd.read_csv(equity_path, index_col=0)
prices = pd.read_csv(prices_path)
```

```python
equity = equity.reset_index().rename(columns={'index':'Date'})
equity['month'] = pd.to_datetime(equity['Date'])
equity['month'] = equity['month'].dt.to_period('M')

equity = equity.groupby(by='month').agg({'Equity':'last'})

equity['perc_diff'] = (equity['Equity'] - equity['Equity'].shift(1)) / equity['Equity'].shift(1)

equity.fillna(0, inplace=True)

equity
```

```python
prices['month'] = pd.to_datetime(prices['Date'])
prices['month'] = prices['month'].dt.to_period('M')

prices = prices.groupby(by='month').agg({'Close':'last'})

prices['perc_diff'] = (prices['Close'] - prices['Close'].shift(1)) / prices['Close'].shift(1)

prices.fillna(0, inplace=True)

prices
```

```python
import plotly.express as px
fig = px.scatter(x=prices['perc_diff'], y=equity['perc_diff'])
fig.update_layout(
    title="Correlacion", 
    xaxis_title='asset monthly returns',
    yaxis_title='strategy monthly returns'
)

fig.show()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

x = np.array(prices['perc_diff']).reshape(-1, 1)
y = equity['perc_diff']

reg = LinearRegression().fit(x, y)
stability_ratio = reg.score(x, y)
stability_ratio
```

```python
import plotly.express as px
import numpy as np
from sklearn.linear_model import LinearRegression

# Datos
x = np.array(prices['perc_diff']).reshape(-1, 1)
y = equity['perc_diff']

# Ajustar el modelo de regresión lineal
reg = LinearRegression().fit(x, y)
stability_ratio = reg.score(x, y)

# Predicciones para la recta
x_range = np.linspace(x.min(), x.max(), 100).reshape(-1, 1)  # Rango de X para la recta
y_pred = reg.predict(x_range)  # Valores predichos de Y

# Crear el gráfico
fig = px.scatter(x=prices['perc_diff'], y=equity['perc_diff'],
                 title="Correlación con regresión lineal",
                 labels={'x': 'Asset Monthly Returns', 'y': 'Strategy Monthly Returns'})

# Agregar la recta de regresión
fig.add_scatter(x=x_range.flatten(), y=y_pred, mode='lines', name='Regresión Lineal')

# Personalización
fig.update_layout(
    xaxis_title='Asset Monthly Returns',
    yaxis_title='Strategy Monthly Returns'
)

# Mostrar
fig.show()
```
