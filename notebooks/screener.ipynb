```python
from finvizfinance.screener.overview import Overview
import pandas as pd
pd.set_option('display.max_columns', None)

```

```python
def get_undervalued_stocks():
    foverview = Overview()
    
    filters_dict = {
        'P/E':'Under 25',
        'Return on Equity': 'Over +15%',
        'Return on Assets' : 'Over +15%',
        'Debt/Equity':'Under 1', 
        'Current Ratio' : 'Over 1',
        'PEG':'Low (<1)', 
        # 'Market Cap.':'+Mid (over $2bln)', 
        # 'Operating Margin':'Positive (>0%)', 
        # 'P/B':'Low (<1)',
        # 'InsiderTransactions':'Positive (>0%)'
    }

    foverview.set_filter(filters_dict=filters_dict)
    df_overview = foverview.screener_view()
    # if not os.path.exists('out'): #ensures you have an 'out' folder ready
    #     os.makedirs('out')
    # df_overview.to_csv('out/Overview.csv', index=False)
    tickers = df_overview['Ticker'].to_list()
    return tickers
```

```python
interest_tickers = get_undervalued_stocks()
interest_tickers
```

```python
columns = [
    'industry',
    'sector',
    'trailingPE',
    'forwardPE',
    'pegRatio',
    'trailingPegRatio'
    'beta',
    'totalDebt',
    'quickRatio',
    'currentRatio',
    'totalRevenue',
    'debtToEquity',
    'revenuePerShare',
    'returnOnAssets',
    'returnOnEquity',
    'freeCashflow',
    'operatingCashflow',
    'earningsGrowth',
    'revenueGrowth',
    'bid',
    'ask',
    'marketCap',
    'twoHundredDayAverage',
    'recommendationKey',
    'numberOfAnalystOpinions',
    'symbol',
]
```

```python
import yfinance as yf

# Diccionario para almacenar la informaci贸n fundamental de cada ticker
fundamental_data = {}

# Obtener la informaci贸n de cada ticker
for ticker in interest_tickers:
    stock = yf.Ticker(ticker)
    info = stock.info  # Informaci贸n fundamental
    fundamental_data[ticker] = info

# Mostrar la informaci贸n fundamental obtenida

metrics = pd.DataFrame()
for ticker, info in fundamental_data.items():
    stock_metrics_dict = {}
    
    for column in columns:
        stock_metrics_dict[column] = [info.get(column)]

    stock_metrics = pd.DataFrame(stock_metrics_dict)
    metrics = pd.concat([metrics, stock_metrics])        
metrics = metrics.set_index('symbol')
```

```python
metrics.sort_values(by='pegRatio', ascending=False)
```
